3.1 Questioning Data and Correlation

Correlation compares how patterns move together between different variables.

Intro to Models: 

The most common model that we run into is the Normal Distribution. It is a bell shaped curve, with the mean in the center. For a normal distribution, about 68% of the values fall within 1 std. deviation and 95% of the values fall within 2 std deviations. 

The Central Limit Thm: as the amount of samples increase, the distribution of sample means approaches a normal distribution regardless of the original shape. Suppose we have a skewed distribution. We randomly select 30 data points of that distibution and get the mean of those 30 data points. Now when we plot the distribution of the sample means, they will folllow a normal dist, and more so as the number of data points continues to increase. 

Vairence and Std Dev: Give us an idea how points drawn from samples of our original distribution differ from the mean of that distribution. 

How to visualize the relationship between two predictors / variables. When both of the variables are numeric we can visualize their relationship with a scatterplot. For example if we plot something like age and study hours, we may see a direct, indirect, or no relationship between the two variables. I.e. we have a single measurement that has data on both of these variables of interest, we can see how these variables interact with one another. 

This relationship between two variables is known as correlation. Correlation primarily tells us two things. 
1) What is the direction of the relationship 
2) What is the strength of the relationship 

Now for Covarience: Covarience is the thing we calculate to get an idea of the direction of the relationship. 

Note though, the magnitude of covarience depends on the units of X and Y. A large positive value doesn't necessarily mean a stronger relationship, it depends on the data scale. To get the relative size we need to scale our covarience to obtain the correlation coefficient. 

The correlation coefficient (r): This is a measure of the strength and direction of the relationship between random variables. It ranges from [-1,1], with greater magnitude indicating a stronger relationship and the sign of r indicating the direction. 

r = 1  - perfect positive linear 
r = -1 - perfect negative linear 
r = 0 - no linear relationship 

Suppose we have a collection of data points on a graph. Most likely comparing two different variables, and we want to try and approximate the true structure of how this data is generated. We can perform a regression, which is a line which minimizes the residuals. The residuals are the measure of the distance between the true data points and the data points which will be predicted by the regresssion line. In another sense, correlation measures how small and consistent the residuals are when comparing the data to its best fit regression line. Like, the data can still have decently big residuals even with its very best fit linear regression, especially in the case where the data may have no relatioship of a nonlinear relationship, so the distance of these residuals gives us an idea about what the magnitue of r should be. 

Hence it makes sense that the correlation coefficient is calculated 

r = Cov(X,Y) / std_dev(X)std_dev(Y)

so we can see that r has both direction, from covarience
and strength, from the std_deviations

Final Remarks: 
As you have probably heard a billion times, correlation does not equal causation. There are many reasons why two variables may appear to move together which is not the true reason we thing. We could have reverse causation, confounding varibales, or just pure coincidence

Linear Regression: 

We are usually focused on seeing the predictive ability of a certain variable on another. 

In almost all cases, we are going to be testing the effect of a predictor(independent variable) on an outcome variable (dependent varaible). 

One of the most simple and common tools in statistical analysis is Linear Regression. Regression is a way we can measure the relationship between two variables. 

Y_hat = b1*X1 + b0 - this is a simple linear regressionn

The goal of a linear regression is to fit a line as best it can to the data. Even the best fit line may stil not be good enough. How do we know if a best fit line is good, we look at the R-Squared value 

Rsquared = 1 - SS_residual / SS_total

where SS_residual is the sum of the differences between the predicted and the actual Y

and SS_total is the sum of the differences between the sample mean and the actual Y

Rsquared can also be thought of as the proportion of the variation in the outcome variable that can be accounted for by the predictor

Note that simple linear regression (one predictorr and one response) is basically the same thing as correlation in many ways

Our regression function performs two statistical tests. 

1) The F-Test: This tests for full model significane. 
We have a Ho (The null hypothesis): none of the variables have an effect on the dependent variable
and then a Ha (Alternative hypothesis): At least one of the variables have an effect on the dependent variable

A significant result (a p-socre less than .05 p<.05) tells us that we can reject the null hypothesis 

If the p_sscore is not less than .05, we fail to reject the null hypothesis (this is not the same thing as accepting the null hypothesis)

Recall for a statistical test, we choose a significance level. The smaller this significance level, the harder it is for us to obtain a significant result because it is more strict. Lover significance level -> higher false negative,  but lower false positive. Depending on the implication of a Type 1 or a Type 2 error a higher/lower significane level be more appropriate.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
```

```python
#here is the dataset of interest
mpg = pd.read_csv("https://raw.githubusercontent.com/Greg-Hallenbeck/HARP-325-DataSociety/main/dataset/mpg-2022.csv", encoding='latin-1')
```

```python
mpg_select = mpg[['City.MPG','Hwy.MPG']]
sns.heatmap(mpg_select.corr())
plt.show() # data leakedge
```

To perfrom a linear regression we need to first preprocess our dataset 

So that means deciding which predictorsto consider as well as converting categorical values (typically we use one-hot-encoding)

Finally, we need to consider data leakage, we need to avoid keeping any variables which secretly contain the correct answer

Our response variable we want to study is the City.MPG, we decide to consider all of the other variables except for Model and Hwy.MPG. The reason why we don't consider Hwy.MPG is because it would be a case of data leakedge. It is likely a simple linear relationship beetween the Hwy.MPG and the City, and it wouldn't be fare to consider it. Because theortically, if we were able to consider Hwy.MPG we could just as easily consider the City.MPG, so we decide to not consider it.Why do we remove model? Well simply because there is 624 different models, many of which only appear once, so we dont want to overfit to the model. Also model may be a case of data leakedge, and one-hot encoding on 624 different models would cause there to be 623 different columns. 

```python
X = mpg[["Displ", "Cyl", "Drive", "Veh.Class", "Air.Pollution.Score", "Greenhouse.Gas.Score", "Comb.CO2"]]
y = mpg["City.MPG"]
```

```python
X = pd.get_dummies(X, drop_first=True).astype("float32")
X = sm.tools.tools.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
print(model.summary())
```

How to read a model summary? 

Adjusted R-squared - tells you approximately what percentage of the variability in the y variable can be explained by the preddicttors. 100% would mean we have a perfect fit. 

The coefficients are the sslopes of how much the y_hat will change wwhen the x-variable is increased by 1. So .63 slope with cylinder means that if holding everything else constant, if we add one more cylinder to the car then we would expect the y-hat to increase .63miles/gallon.

The coef for the categorical variables repressents the change in y_hat we obtain when switching from the reference category to this categorty. So when it says a 4WD has a .33, that means that compared to a 2WD car, a 4WD car has a gas mileage of .33mile/gallon highter. 

standard error: this column iss the uncertainty on the intercept and the slope

P > |t| - the statistical significance, for each of the slope and intercept combos of the predictors, python produces a respective p-value which tells us the probability that this predicted slope has explanatory power.

```python
#Now for model evaluation 
yhat = model.predict()
residuals = y - yhat
```

```python
residuals
ax = sns.scatterplot(x=yhat, y=residuals)
ax.axhline(y=0, color='black', linestyle='dashed')

plt.xlabel("Predicted MPG")
plt.ylabel("Residual")
plt.show()
```

How do we evaluate  the residuals, we want to see if there  is a trend in the residuals graph which would imply that our model is betweer at making predictions for certain miles per gallon vs at others. Ideally we want to see the residuals randomly distributed and cloe to a straight line at residual value of 0. The closer our residduals are to 0, the lower the bias is in making predictions. 

As we can see by the residuals chart we made, we do a much better job at predicting the MPG when dealing with more mid range MPG predictions. For whatever reason higher MPG vehicles dont follow the same variable relationships as mid range vehicles.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
```

```python
bmd = pd.read_csv("https://www.dropbox.com/s/c6mhgatkotuze8o/bmd.csv?dl=1")
```

The response variable is fracture
The predictors are age, sex, weight, height, medication, waiting time, and bmd 
We will remove the id column

```python
response = bmd.fracture
response = pd.get_dummies(response, drop_first = True, ).astype('float32')

#be very carful to note here 1 = no fracture, and  0 = fracture

bmd_predictors = bmd.drop(columns = ['id','fracture'])
bmd_predictors = pd.get_dummies(bmd_predictors, drop_first=True).astype('float32')

#notice for some of these categorial variabes it is redundant to keep both
#for example keeping sex_F and sex_M when they are representing the same data is silly

# bmd_predictors = bmd_predictors.drop(columns = ['sex_M','medication_No medication'])

#also going to add a bmi column

bmd_predictors['bmi'] = bmd_predictors['weight_kg'] / (bmd_predictors['height_cm']/100)**2
```

We obtained an R^2 of .45 which is pretty good. So just these few variables tell us 45%. We found age and bmd to be statisticall significant at different levels. 

The sigmoid function: 

S = 1 / (1 + exp(-Input))

Basically we can calcularte  the predicted probability, 

p_hat  = 1 / (1 + exp(-z_hat))

where z_hat is the linear model we get

```python
bmd_predictors = sm.tools.tools.add_constant(bmd_predictors)

model = sm.Logit(response, bmd_predictors).fit()

print(model.summary())
```

How do we interpret this model? 

we have produced a linear model again, and the output of this model is what we will feed to the sigmoid function. 

Our coeficients tells uus the expected changed in log-odds 

So looking at the height variable, increasing height by 1-cm causes an a change in odds-ratio of exp(.1281) = 1.137

so noww to get the odds percentage, we do 

(1.137 - 1) x 100% = 13.7%

```python
phat = model.predict()

phat[0:10]

#a prob closer to 1 means the patient is less likely 
# to experience a fracture while a prob closer to 0 means they are more likely
```
