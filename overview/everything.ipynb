3.1 Questioning Data and Correlation

Correlation compares how patterns move together between different variables.

Intro to Models: 

The most common model that we run into is the Normal Distribution. It is a bell shaped curve, with the mean in the center. For a normal distribution, about 68% of the values fall within 1 std. deviation and 95% of the values fall within 2 std deviations. 

The Central Limit Thm: as the amount of samples increase, the distribution of sample means approaches a normal distribution regardless of the original shape. Suppose we have a skewed distribution. We randomly select 30 data points of that distibution and get the mean of those 30 data points. Now when we plot the distribution of the sample means, they will folllow a normal dist, and more so as the number of data points continues to increase. 

Vairence and Std Dev: Give us an idea how points drawn from samples of our original distribution differ from the mean of that distribution. 

How to visualize the relationship between two predictors / variables. When both of the variables are numeric we can visualize their relationship with a scatterplot. For example if we plot something like age and study hours, we may see a direct, indirect, or no relationship between the two variables. I.e. we have a single measurement that has data on both of these variables of interest, we can see how these variables interact with one another. 

This relationship between two variables is known as correlation. Correlation primarily tells us two things. 
1) What is the direction of the relationship 
2) What is the strength of the relationship 

Now for Covarience: Covarience is the thing we calculate to get an idea of the direction of the relationship. 

Note though, the magnitude of covarience depends on the units of X and Y. A large positive value doesn't necessarily mean a stronger relationship, it depends on the data scale. To get the relative size we need to scale our covarience to obtain the correlation coefficient. 

The correlation coefficient (r): This is a measure of the strength and direction of the relationship between random variables. It ranges from [-1,1], with greater magnitude indicating a stronger relationship and the sign of r indicating the direction. 

r = 1  - perfect positive linear 
r = -1 - perfect negative linear 
r = 0 - no linear relationship 

Suppose we have a collection of data points on a graph. Most likely comparing two different variables, and we want to try and approximate the true structure of how this data is generated. We can perform a regression, which is a line which minimizes the residuals. The residuals are the measure of the distance between the true data points and the data points which will be predicted by the regresssion line. In another sense, correlation measures how small and consistent the residuals are when comparing the data to its best fit regression line. Like, the data can still have decently big residuals even with its very best fit linear regression, especially in the case where the data may have no relatioship of a nonlinear relationship, so the distance of these residuals gives us an idea about what the magnitue of r should be. 

Hence it makes sense that the correlation coefficient is calculated 

r = Cov(X,Y) / std_dev(X)std_dev(Y)

so we can see that r has both direction, from covarience
and strength, from the std_deviations

Final Remarks: 
As you have probably heard a billion times, correlation does not equal causation. There are many reasons why two variables may appear to move together which is not the true reason we thing. We could have reverse causation, confounding varibales, or just pure coincidence
