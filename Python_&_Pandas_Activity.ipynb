## Return on Investment
As a starting question, I'd like to re-do some of the analysis done by [this FiveThirtyEight article](https://fivethirtyeight.com/features/the-dollar-and-cents-case-against-hollywoods-exclusion-of-women/): do movies which pass the Bechdel test have a higher return on investment than those that don't? We'll take a couple of steps to get there.

Unfortunately, if you take a peek at the result of `head(bechdel)` the budget columns are characters, not numeric! That's because some of the entries don't have a listed budget or box office take, instead being listed as "#N/A". So, let's remove these.

First, only keep the `year`, `title`, `clean_test`, `budget_2013$`, `intgross_2013$`, and `domgross_2013$` columns. Save the result in a variable caled `profits`.

```python
bechdel_subset = bechdel[['year','title','clean_test','budget_2013$','intgross_2013$','domgross_2013$']]
```

538 only looked at films after 1990, do let's so the same. Save the result as our profit variable

```python
profits = bechdel_subset[bechdel_subset.year > 1990]
```

Because typing the `_2013$` at the end of every name is a pain, let's rename() to remove that part so they just have the names `budget`, and so on. Save it in `profits`.

```python
profits = profits.rename(columns = {'budget_2013$':"budget_2013",
                          "domgross_2013$":"domgross_2013",
                          "intgross_2013$":"intgross"})
```


Finally, let's calculate the domestic return on investment, or `roi`, which is the domestic gross divided by the budget of the film. We'll need to add a column to do this.

```python
profits['roi'] = profits.domgross_2013 / profits.budget_2013
profits
```

Now we're ready to actually summarize the data and get the results we want!
Use `group_by()` to calculate the `mean()` return on investment for each different test result. You can also sort by the `roi` to see them in order a little more clearly. What trends do you see?

```python
profits.groupby('clean_test')['roi'].mean()
```
